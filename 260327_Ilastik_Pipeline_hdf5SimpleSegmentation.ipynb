```python
#Modules
from pathlib import Path
import subprocess
import numpy as np
import pandas as pd
import h5py
import subprocess
import os
```

### IMPORTANT

```python
# ── User-defined parameters ────────────────────────────────────────────────────
# Root directory containing all mouse sub-folders with TIFF images
basepath = "E:\\260306_AxonaThioS_Tiffs\\"

# List of mouse IDs to process; one sub-folder per ID is expected under basepath
mouse_id = [ "644788", "708153","726857", "733075", "1116806", "1116808", "1116810", "1117201", "1117425"]
```

```python

# ── Ilastik Pixel Classification: TIFF → Probability Maps (HDF5) ──────────────
# For each mouse, this cell:
#   1. Reads all .tif ROI images from the "ROIs" sub-folder
#   2. Runs ilastik in headless mode to generate per-pixel probability maps
#   3. Saves the output as HDF5 files in "Ilastik_Pipeline_Outputs"
#
# NOTE: Update the --project path if using a different ilastik project file
#       (e.g. switch 'NLGF' to 'NLF' for a different genotype cohort).

for mouse in mouse_id:
    in_folder  = Path(basepath) / mouse / "ROIs"
    out_folder = Path(basepath) / mouse / "Ilastik_Pipeline_Outputs"

    # Create output folder if it doesn't exist
    out_folder.mkdir(parents=True, exist_ok=True)

    # Override MPLBACKEND so ilastik doesn't inherit Jupyter's inline backend,
    # which would cause a crash when running headless
    env = os.environ.copy()
    env['MPLBACKEND'] = 'Agg'

    # ── Process each TIFF file ─────────────────────────────────────────────────
    for in_file in in_folder.glob("*.tif"):
        # Output HDF5 probability map mirrors the input filename
        out_file = out_folder / in_file.name.replace(".tif", "_Probabilities.h5") # Change suffix to indicate simple segmentation output

        print(f"Processing {in_file} -> {out_file}")

        # Build the ilastik headless command
        command = [
            "C:\\Program Files\\ilastik-1.4.2b6\\ilastik.exe",
            '--headless',
            '--project=E:\\2603 Diego Ilastik Probability Projects\\260323_NLGF_Axona_Pixel_Classification_Probabilites.ilp',
            '--output_format=hdf5',
            '--export_source=probabilities',   # export class probability maps, option to also export simple segmentation if desired
            '--raw_data=' + str(in_file),
            '--output_filename_format=' + str(out_file)
        ]

        subprocess.run(command, env=env, check=True)
        print("Done.")
```

```python

# ── Ilastik Pixel Classification: TIFF → Simple Segmentation (HDF5) ──────────────
# For each mouse, this cell:
#   1. Reads all .tif ROI images from the "ROIs" sub-folder
#   2. Runs ilastik in headless mode to generate per-pixel simple segmentation maps
#   3. Saves the output as HDF5 files in "Ilastik_Pipeline_Outputs"
#
# NOTE: Update the --project path if using a different ilastik project file
#       (e.g. switch 'NLGF' to 'NLF' for a different genotype cohort).

for mouse in mouse_id:
    in_folder  = Path(basepath) / mouse / "ROIs"
    out_folder = Path(basepath) / mouse / "Ilastik_Pipeline_Outputs"

    # Create output folder if it doesn't exist
    out_folder.mkdir(parents=True, exist_ok=True)

    # Override MPLBACKEND so ilastik doesn't inherit Jupyter's inline backend,
    # which would cause a crash when running headless
    env = os.environ.copy()
    env['MPLBACKEND'] = 'Agg'

    # ── Process each TIFF file ─────────────────────────────────────────────────
    for in_file in in_folder.glob("*.tif"):
        # Output HDF5 probability map mirrors the input filename
        out_file = out_folder / in_file.name.replace(".tif", "_SimpleSegmentation.h5") # Change suffix to indicate simple segmentation output

        print(f"Processing {in_file} -> {out_file}")

        # Build the ilastik headless command
        command = [
            "C:\\Program Files\\ilastik-1.4.2b6\\ilastik.exe",
            '--headless',
            '--project=E:\\2603 Diego Ilastik Probability Projects\\260323_NLGF_Axona_Pixel_Classification_Probabilites.ilp',
            '--output_format=hdf5',
            '--export_source=simple segmentation',   # export class simple segmentation maps, option to also export probabilities if desired
            '--raw_data=' + str(in_file),
            '--output_filename_format=' + str(out_file)
        ]

        subprocess.run(command, env=env, check=True)
        
        print("Done.")
```

```python

# ── HDF5 Simple Segmentation Quantification→ Percentage Coverage ────────────────────────────
# For each mouse, this cell:
#   1. Finds all *_SimpleSegmentation.h5 files exported by ilastik
#   2. Parses the filename to extract metadata (staining, genotype, gender, etc.)
#   3. Counts plaque vs. background pixels and computes % coverage
#   4. Saves one CSV per unique brain/region combination

for mouse in mouse_id:
    out_folder = Path(basepath) / mouse / "Ilastik_Pipeline_Outputs"

    # Collect all SimpleSegmentation HDF5 files for this mouse
    hdf5_files = out_folder.glob("*_SimpleSegmentation.h5")

    # Accumulates per-file results, keyed by a unique metadata string
    hdf5_dict = {}

    # ── Process each segmentation file ────────────────────────────────────────
    for in_file in hdf5_files:
        filename = in_file.stem
        # Expected filename format:
        # <StainingMethod>_<Genotype>_<Gender>_<Brain>_<Slide>_<Section>_<channel>_<Region>_SimpleSegmentation
        parts = filename.split("_")

        # Skip files that don't match the expected naming convention
        if len(parts) >= 7:
            staining_method = parts[0]
            genotype        = parts[1]
            gender          = parts[2]
            brain           = parts[3]
            slide           = parts[4]
            section         = parts[5]
            region          = parts[-1]  # region is the last part before "_SimpleSegmentation"

            print(f"Processing: {in_file.name}")

            # ── Read segmentation array from HDF5 ─────────────────────────────
            with h5py.File(in_file, 'r') as f:
                # ilastik can export under different dataset names; handle both
                if 'exported_data' in f:
                    segmentation_data = f['exported_data'][:]
                elif 'data' in f:
                    segmentation_data = f['data'][:]
                else:
                    print(f"⚠️ Skipping {in_file.name} - unknown structure: {list(f.keys())}")
                    continue

                # Remove singleton dimensions added by ilastik (e.g. z-axis of 1)
                segmentation_data = np.squeeze(segmentation_data)

                # ── Pixel counting ────────────────────────────────────────────
                # Label values as defined in the ilastik project
                plaque_label     = 3  # amyloid plaques
                parenchyma_label = 2  # tissue background (excludes label 1 = background outside tissue)

                plaque_count     = np.count_nonzero(segmentation_data == plaque_label)
                parenchyma_count = np.count_nonzero(segmentation_data == parenchyma_label)

                # % coverage = plaques / (plaques + parenchyma) × 100
                if (plaque_count + parenchyma_count) == 0:
                    percentage_coverage = 0  # no tissue pixels found; set to 0
                else:
                    percentage_coverage = (plaque_count / (plaque_count + parenchyma_count)) * 100

                # ── Store result ──────────────────────────────────────────────
                # Group slides/sections belonging to the same brain & region
                unique_key = f"{staining_method}_{genotype}_{gender}_{brain}_{region}"

                if unique_key not in hdf5_dict:
                    hdf5_dict[unique_key] = {'slide': [], 'section': [], 'percentage_coverage': []}

                hdf5_dict[unique_key]['slide'].append(slide)
                hdf5_dict[unique_key]['section'].append(section)
                hdf5_dict[unique_key]['percentage_coverage'].append(percentage_coverage)

    # ── Export one CSV per unique brain/region combination ────────────────────
    for key, value in hdf5_dict.items():
        slides    = value['slide']
        sections  = value['section']
        coverages = value['percentage_coverage']

        # Append a summary row with the mean coverage across all sections
        average = sum(coverages) / len(coverages)
        slides.append("AVERAGE")
        sections.append("")
        coverages.append(average)

        df = pd.DataFrame({
            "Slide":               slides,
            "Section":             sections,
            "Percentage Coverage": coverages
        })

        output_filename = f"{key}_Percentage_Of_Coverage.csv"
        output_path     = out_folder / output_filename

        df.to_csv(output_path, index=False)
        print(f"Saved: {output_filename}")
```

```python
#Quantification of plaque classification
csv_files = out_folder.glob("*.csv")

csv_dict = {}

for in_file in csv_files:
    filename = in_file.stem
    parts = filename.split("_")
    
    if len(parts) >= 6:
        staining_method = parts[0]
        genotype = parts[1]
        gender = parts[2]
        brain = parts[3]
        slide = parts[4]
        region = parts[5]
        
        unique_key = f"{brain}_{region}"
        
        if unique_key not in csv_dict:
            csv_dict[unique_key] = {'data': []}
        
        read = pd.read_csv(in_file)
        csv_dict[unique_key]['data'].append(read)

for key, value in csv_dict.items():
    data = pd.concat(value['data'])
    c = data['Predicted Class'].value_counts(normalize=True) * 100
    
    class_dist_row = pd.DataFrame([c], columns=c.index)
    result = pd.concat([class_dist_row, data], ignore_index=True)
    
    output_filename = f"{key}_SML.xlsx"
    output_path = out_folder / output_filename
    
    writer = pd.ExcelWriter(output_path, engine='xlsxwriter')
    result.to_excel(writer, sheet_name='SML', index=False)
    writer.close()
```
